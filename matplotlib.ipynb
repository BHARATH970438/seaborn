Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Note: Also add title, xlabel, and ylabel to the plot.

Scatter Plot: A scatter plot represents individual data points as markers on a graph. It is useful for visualizing the relationship between two numerical variables and identifying any patterns or correlations.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x = x, y = y)
plt.title('This is Title')
plt.xlabel('x_label')
plt.ylabel('Y_label')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
line_1_x = np.array([0, 1, 2, 3, 4, 5])
line_1_y = np.array([0, 100, 200, 300, 400, 500])
line_2_x = np.array([0, 1, 2, 3, 4, 5])
line_2_y = np.array([50, 20, 40, 20, 60, 70])
line_3_x = np.array([0, 1, 2, 3, 4, 5])
line_3_y = np.array([10, 20, 30, 40, 50, 60])
line_4_x = np.array([0, 1, 2, 3, 4, 5])
line_4_y = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt

plt.subplot(2,2,1)
plt.plot(line_1_x,line_1_y)
plt.title('Graph_1')

plt.subplot(2,2,2)
plt.plot(line_2_x,line_2_y)
plt.title('Graph_2')

plt.subplot(2,2,3)
plt.plot(line_3_x,line_3_y)
plt.title('Graph_3')

plt.subplot(2,2,4)
plt.plot(line_4_x,line_4_y)
plt.title('Graph_4')

plt.tight_layout()  # To adjust spacing between subplots.
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar Plot: A bar plot displays categorical data as rectangular bars. It is commonly used to compare the values of different categories and show their relative magnitudes.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt

plt.bar(x = company, height = profit)
plt.title('Company_Profit')
plt.xlabel('Companies')
plt.ylabel('Profit')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

a box plot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset. It provides a visual summary of the key statistical measures such as the median, quartiles, and outliers. Box plots are commonly used to analyze and compare the distributions of different variables or groups.

```python
import numpy as numpy

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(x = [box1,box2])

plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Box Plot')

plt.show()
```
