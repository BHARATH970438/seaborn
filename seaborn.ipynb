Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

The Seaborn library is a popular data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics.
- Here five commonly used plots :
1. Line Plot: A line plot displays data points connected by straight lines, making it useful for visualizing trends or changes over time. It is commonly used to depict continuous data, such as stock prices, temperature variations, or population growth.

2. Bar Plot: A bar plot shows categorical data as rectangular bars, with the height of each bar representing a specific value. It is commonly used to compare different categories or groups and is suitable for displaying counts, frequencies, or average values.

3. Histogram: A histogram represents the distribution of a continuous variable by dividing it into bins and displaying the frequency or count of data points falling into each bin. It is useful for visualizing data distribution, detecting outliers, and understanding the shape of the data.
4. Scatter Plot: A scatter plot displays individual data points as dots on a two-dimensional plane, with one variable on the x-axis and another on the y-axis. It is effective for exploring relationships and patterns between two continuous variables. Scatter plots can also be enhanced with additional information, such as color or size, to represent additional dimensions.

5. Heatmap: A heatmap visualizes data using a color-coded matrix, where different colors represent different values. It is often used to show the correlation or relationship between variables in a tabular format. Heatmaps are particularly useful when working with large datasets and can provide insights into patterns and clusters.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using
- x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plty

data = sns.load_dataset('titanic')
data.head(3)
```

```python
sns.boxplot(x= 'pclass',y='age',data= data).set_title(' Age by Passenger Class ')
plty.show()

sns.boxplot(x= 'pclass',y='fare',data= data).set_title(' Fare by Passenger Class ')
plty.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('flights')
data.head(3)
```

```python
flights_pivot = data.pivot_table(values="passengers", index="month", columns="year")

plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt=".1f")
plt.title("Number of flights by month and year")
plt.show()
```
